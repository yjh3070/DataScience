```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/suggestion.csv', encoding='euc-kr')
#df.shape
df.head()
```

```python
# 열을 drop 할 때는 axis값은 1
df.drop(['Unnamed: 0'], inplace=True, axis=1)
# del df['Unnamed: 0']

# 행을 drop 할 때는 axis값은 0이고 0은 기본 값으로 있어서 생략가능
# df.drop([0], inplace=True, axis=0)
df.head()
```

```python
# 내가 원하는 열을 index로 지정
df.set_index('sgId', inplace=True)
df.head()
```

```python
df['성별'] = 0
df['a']=1
df.head()
```

```python
df.drop(['성별'], inplace=True, axis=1)
del df['a']
df.head()
```
