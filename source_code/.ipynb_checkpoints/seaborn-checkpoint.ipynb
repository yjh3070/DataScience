```python
# 라이브러리 임포트하기

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = 'darkgrid')

%matplotlib inline
```

```python
# 데이터셋

tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
#hue 색상
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
#hue 색상
sns.relplot(x='total_bill', y='tip', data=tips, hue='size')
```

```python
#style & hue
sns.relplot(x='total_bill', y='tip', data=tips, style='smoker', hue='size')
```

```python
#색상
sns.relplot(x='total_bill', y='tip', data=tips,  hue='size', palette='Blues')
```

```python
#색상
sns.relplot(x='total_bill', y='tip', data=tips,  size='size')
```

```python
df = pd.DataFrame(dict(time=np.arange(500), value = np.random.randn(500).cumsum()))
df
```

```python
sns.relplot(x='time', y='value', data=df, kind='line')
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x='timepoint', y='signal', kind = 'line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal', kind = 'line', data=fmri, ci=None)
```

```python
sns.relplot(x='timepoint', y='signal', kind = 'line', data=fmri, ci='sd')
```

```python
# cat plot
tips
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
# cat plot
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

```python
# cat plot
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
# cat plot
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex', split=True)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', markers=['s', '*'], height=7)
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot(index = 'month', columns = 'year', values = 'passengers')
flights.head()
```

```python
sns.heatmap(flights, annot=True, fmt='d', linewidths=.5)
```

```python
sns.set_style=('tips')
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris, hue='species')
```

```python
sns.pairplot(iris, hue='species', vars=['sepal_width', 'sepal_length'])
```

```python
sns.pairplot(iris, x_vars=['sepal_width', 'sepal_length'], y_vars=['petal_width', 'petal_length'], hue='species')
```

```python
sns.pairplot(iris, kind='reg', size=3)
```
