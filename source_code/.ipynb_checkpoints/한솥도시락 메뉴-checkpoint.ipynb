```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
driver = wb.Chrome()

url = "https://www.hsd.co.kr/menu/menu_list"

driver.get(url)

btn = driver.find_element_by_css_selector('.c_05');

for i in range(3) :
    btn.click()
    time.sleep(2)
```

```python
# 모든 도시락 정보가 나온 경우 bs4로 변환
soup = bs(driver.page_source, 'html.parser')
# lxml, html.parser
```

```python
name_list = []
name_tag_list = soup.find_all('h4', class_ = "h")
for name_tag in name_tag_list :
    name_list.append(name_tag.string)
name_list
```

```python
price_list = []
price_tag_list = soup.select('.item-price strong')
for price_tag in price_tag_list :
    price_list.append(price_tag.string)
price_list
```

```python
dic = {'name' : name_list, 'price' : price_list}
data = pd.DataFrame(dic)
data
```

```python
data.to_csv('../data/한솥(euc-kr).csv', encoding='euc-kr', index=False)
```

```python
data.to_csv('../data/한솥(utf-8).csv', encoding='utf-8', index=False)
```
