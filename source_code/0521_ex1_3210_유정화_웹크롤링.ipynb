```python
def get(url):
    article = []

    baseurl = f"http://www.innogov.go.kr{url}"
    html = urlopen(baseurl)
    soup = BeautifulSoup(html, "html.parser")
            
    sgId = url.split("=")[2].split("&")[0]
    article.append(sgId)

    title = soup.select_one("#content > div.vveiw_box1 > dl > dt").string
    article.append(title)
    
    data = soup.select('li > dl > dd')
    
    category = data[0].string
    start = data[1].string
    end = data[2].string
    author = data[3].string

#     category = soup.select_one("#content > div.vveiw_box1 > div.vveiw_name > ul > li:nth-child(1) > dl > dd").string
    article.append(category)

#     start = soup.select_one("#content > div.vveiw_box1 > div.vveiw_name > ul > li:nth-child(2) > dl > dd").string
    article.append(start)

#     end = soup.select_one("#content > div.vveiw_box1 > div.vveiw_name > ul > li:nth-child(3) > dl > dd").string
    article.append(end)

#     author = soup.select_one("#content > div.vveiw_box1 > div.vveiw_name > ul > li:nth-child(4) > dl > dd").string
    article.append(author)
    
    vote = soup.select_one("#counter").string
    article.append(vote)

    return article
```
