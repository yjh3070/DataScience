```python
t = np.arange(0, 10, 0.1)
y = np.sin(t)
```

```python
# .plot(x축, y축)
plt.plot(t, y, color = 'cyan', linestyle='--', linewidth= 5, marker='')
plt.show()
```

```python
# 하나의 리스트나 어레이를 넣으면 y축을 기준으로 값이 나온다.
plt.plot([1,2,3,4,5])
plt.show()
```

```python
# 창의 크기 변경
plt.figure(figsize=(6,7))
# 사인함수 그래프
plt.plot(t, y, color = 'cyan', linestyle='--', linewidth= 5, marker='', label='sin')
# 코사인함수 그래프
plt.plot(t, np.cos(t), color='r', linestyle='-', linewidth=5, label='cos')

# 그래프 옵션들
# 격자무늬
plt.grid()

# 제목
plt.title('사인함수, 코사인함수')

# x축 이름
plt.xlabel('x축')

# y축 이름
plt.ylabel('y축')

# y축 범위 (x축 범위도 똑같이 .xlim)
plt.ylim(-2, 2)

# 글자 입력 .text(x값, y값, '내용')
plt.text(6, 1.5, '빅데이터분석')

# 범례
plt.legend()

plt.show()
```

```python
# .subplot(행, 열, 번째)
# 1행 2열의 그래프를 그릴 건데 그 중 1번째
plt.subplot(1,2,1)
plt.plot(t, y)

plt.subplot(1,2,2)
plt.plot(t, np.cos(y))

plt.show()
```
